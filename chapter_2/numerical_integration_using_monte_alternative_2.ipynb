```python
import numpy as np
import matplotlib.pyplot as plt
# https://youtu.be/EnsWVUjpqDE
```

```python
a = 2
b = 5
n = 10000
```

```python
x = np.linspace(a, b , n)
y = x**3
```

```python
plt.title("$f(x) = x^3$")
plt.plot(x, y)
```

```python
def f(x):
    return (x**3)
```

```python
X = np.random.uniform(a, b , n)
Y = f(X)*(b-a)
```

```python
ans = np.sum(Y)/n
ans
```
